# Tournament xG Analysis - WWC 2023

This notebook is lesson 6 of the Soccermatics course by David Sumpter.

**Goals:**
- Load all 64 matches from the Women's World Cup 2023
- Calculate xG for every team across the tournament
- Build a tournament table ranked by xG performance
- Compare xG ranking vs actual points ranking

**Libraries used:** `statsbombpy`, `pandas`, `matplotlib`

```python
#Loading Imports
from statsbombpy import sb
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Laod all 64 matches from the Women's World Cup 2023
matches = sb.matches(competition_id=72, season_id=107)
print(f"Total matches: {len(matches)}")
matches[['match_id', 'home_team', 'away_team', 'home_score', 'away_score']].head()
```

```python
#Loop that goes through each match and calculates the xG for each team
match_ids = matches['match_id'].tolist()

all_xg = []

for match_id in match_ids:
    events = sb.events(match_id=match_id)
    shots = events[events['type'] == 'Shot']
    
    for team, group in shots.groupby('team'):
        all_xg.append({
            'match_id': match_id,
            'team': team,
            'xg': group['shot_statsbomb_xg'].sum()
        })

xg_df = pd.DataFrame(all_xg)
xg_df.head(10)
```

```python
#Group by team and sum the xG for each team
tournament_xg = xg_df.groupby('team')['xg'].sum().reset_index()
tournament_xg.columns = ['team', 'total_xg']
tournament_xg = tournament_xg.sort_values('total_xg', ascending=False).reset_index(drop=True)
tournament_xg['total_xg'] = tournament_xg['total_xg'].round(2)
tournament_xg.head(10)
```

```python
#Add a rank column to the tournament_xg dataframe
tournament_xg['xg_rank'] = tournament_xg.index + 1
print(tournament_xg[['team', 'total_xg', 'xg_rank']].to_string())
```
