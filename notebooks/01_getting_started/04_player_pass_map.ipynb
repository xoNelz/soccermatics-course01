# Player Pass Map and Heatmap

This notebook is lesson 4 of the Soccermatics course by David Sumpter.

**Goals:**
- Filter passes down to a single player
- Build a pass map for that player
- Create a heatmap showing their positional activity
- Tell a player story through data visualization

**Libraries used:** `statsbombpy`, `mplsoccer`, `pandas`, `matplotlib`

```python
from statsbombpy import sb
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer import Pitch, VerticalPitch
```

```python
#Number of passes by Nigerian players
events = sb.events(match_id=3893790)
passes = events[events['type'] == 'Pass']
nigeria_passes = passes[passes['team'] == 'Nigeria Women\'s'].copy()

nigeria_passes['player'].value_counts()
```

```python
#Filtering Deborah Abiodun's passes
player_passes = nigeria_passes[nigeria_passes['player'] == 'Deborah Ajibola Abiodun'].copy()

player_passes['x'] = player_passes['location'].apply(lambda loc: loc[0])
player_passes['y'] = player_passes['location'].apply(lambda loc: loc[1])
player_passes['end_x'] = player_passes['pass_end_location'].apply(lambda loc: loc[0])
player_passes['end_y'] = player_passes['pass_end_location'].apply(lambda loc: loc[1])

player_passes[['x', 'y', 'end_x', 'end_y', 'pass_outcome']].head()
```

```python
#Deborah Abiodun's Pass Map
pitch = Pitch(pitch_type='statsbomb', pitch_color='grass', line_color='white')
fig, ax = pitch.draw(figsize=(12, 8))

for _, row in player_passes.iterrows():
    if row['pass_outcome'] != row['pass_outcome']:  # NaN = completed
        color = 'white'
    else:
        color = 'red'
    
    pitch.arrows(row['x'], row['y'], row['end_x'], row['end_y'],
                 ax=ax, color=color, width=2, headwidth=5)

ax.set_title("Deborah Ajibola Abiodun - Pass Map vs Canada - WWC 2023", 
             color='white', fontsize=14)
fig.patch.set_facecolor('green')

plt.show()
```

```python
#Deborah Abiodun's Heatmap
pitch = Pitch(pitch_type='statsbomb', pitch_color='grass', line_color='white')
fig, ax = pitch.draw(figsize=(12, 8))

pitch.kdeplot(player_passes['x'], player_passes['y'], 
              ax=ax, fill=True, cmap='hot', alpha=0.5)

ax.set_title("Deborah Ajibola Abiodun - Heatmap vs Canada - WWC 2023", 
             color='white', fontsize=14)
fig.patch.set_facecolor('green')

plt.show()
```
