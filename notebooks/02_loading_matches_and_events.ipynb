**Goals:**
- Select a competition and season from StatsBomb open data

**Libraries used:** `statsbombpy`, `pandas`

```python
from statsbombpy import sb
import pandas as pd
```

```python
matches = sb.matches(competition_id=72, season_id=107)
matches.shape
```

```python
matches.columns.tolist()
```

```python
matches[['match_id', 'match_date', 'home_team', 'away_team', 'home_score', 'away_score', 'competition_stage']].head(10)
```

```python
events = sb.events(match_id=3893790)
events.shape
```

```python
events['type'].value_counts()
```

```python
events[['minute', 'second', 'team', 'player', 'type', 'location']].head(10)
```
